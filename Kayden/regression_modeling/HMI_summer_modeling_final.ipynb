```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
# Filtering dataset which does not include NA value in both variables(wellbeing_t2_m, wellbeing_t0_m)
data_no_na = data.dropna(subset=[
    'wellbeing_t2_m',
    'wellbeing_t0_m',
    'pre_meditation_frequency',
    'cv_daily_elapsed_time',
    'cv_interval',
    'lcs_score_sum'])
# Create output variable (wellbeing_t2_m - wellbeing_t0_m)
#data_no_na['wellbeing_diff'] = data_no_na['wellbeing_t2_m'] - data_no_na['wellbeing_t0_m']
```

```python
sns.heatmap(data.isnull().transpose())
```

```python
data['awareness_change'] = data["awareness_t1_m"]-data["awareness_t0_m"]
```

```python
# Create linear regression model
results = smf.ols('mental_wellbeing_change ~ stress_total + days_used + total_elapsed_time + total_session_completed + cv_daily_elapsed_time + cv_interval + lcs_score_sum', data=data).fit()
```

```python
# output variable = change of awareness scores
results2 = smf.ols('awareness_change ~ stress_total + days_used + total_elapsed_time + total_session_completed + cv_daily_elapsed_time + cv_interval + lcs_score_sum', data=data).fit()
```

```python
print(results.summary())
```

```python
print(results2.summary())
```

```python
# Add one more parameter(pre-meditation frequency) to this model

# First, reduct dimenstionality of pre-meditation frequency from 7 to 3.
data = data[data['pre_meditation_frequency']!='prefer_not_to_say']
data['pre_med_reduc'] = data['pre_meditation_frequency'].replace({
    '7_or_more' : "Frequently",
    '5-6' : "Frequently",
    'less_than_once' : "Sometimes", 
    '3-4' : "Sometimes", 
    '1-2' : "Sometimes", 
    'didnt_meditate' : "Never"})

# Use POC Contrast Coding

## pre meditation experience vs no meditation experience
group_map_1 = {
    "Frequently": 0.333,
    "Sometimes" : 0.333,
    "Never" : -0.667
}

## Frequently vs Sometimes
group_map_2 = {
    "Frequently": 0.5,
    "Sometimes" : -0.5,
    "Never" : 0
}

data['contrast1'] = data['pre_med_reduc'].map(group_map_1)
data['contrast2'] = data['pre_med_reduc'].map(group_map_2)
```

```python
# check the multicollinearity
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant

cols = ["cv_interval", "cv_daily_elapsed_time", "stress_total", "days_used", "total_elapsed_time", "total_session_completed", "lcs_score_sum"
]

X = data[cols]
X = add_constant(X)

X = X.dropna()

vif_df = pd.DataFrame({
    "feature": X.columns,
    "VIF": [variance_inflation_factor(X.values, i)
            for i in range(X.shape[1])]
})
print(vif_df)
```

```python
# regression (drop total_elapsed_time, total_session_completed variables)
results1 = smf.ols('mental_wellbeing_change ~ stress_total + contrast1 + contrast2 + days_used + cv_daily_elapsed_time + cv_interval + lcs_score_sum', data=data).fit()
```

```python
print(results1.summary())
```
